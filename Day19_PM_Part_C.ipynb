Q1 — Difference between PMF and PDF

PMF (Probability Mass Function) is used for discrete random variables. It gives the probability of exact values. Example: probability of getting exactly 3 heads in 5 coin tosses.

PDF (Probability Density Function) is used for continuous random variables. It does not give exact probability at a point but describes the density over an interval. Example: distribution of heights or weights.

Key difference:

PMF → discrete values, probabilities sum to 1

PDF → continuous values, area under curve = 1

Q2 — Function to compute probability from frequency

```python
def probability_from_frequency(data, event):
    total = len(data)
    count = sum(1 for x in data if x == event)
    return count / total


# Example
data = [1, 2, 2, 3, 3, 3, 4]
print(probability_from_frequency(data, 3))
```

Q3 — Central Limit Theorem (CLT)

The Central Limit Theorem states that the distribution of sample means becomes approximately normal as the sample size increases, regardless of the original data distribution.

It is important in machine learning because:

It allows us to assume normality for large samples

It helps in hypothesis testing and confidence intervals

Many algorithms rely on normal approximation for inference
