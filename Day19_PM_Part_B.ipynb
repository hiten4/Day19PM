```python
import numpy as np
import matplotlib.pyplot as plt
```


# 1. Generate samples

```python
np.random.seed(42)

normal_data = np.random.normal(loc=0, scale=1, size=1000)
uniform_data = np.random.uniform(low=0, high=1, size=1000)
```



# 2. Plot distributions

```python
plt.figure()

plt.hist(normal_data, bins=30, alpha=0.6, label="Normal")
plt.hist(uniform_data, bins=30, alpha=0.6, label="Uniform")

plt.title("Normal vs Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

# 3. Compare statistics

```python
normal_mean = np.mean(normal_data)
uniform_mean = np.mean(uniform_data)

normal_std = np.std(normal_data)
uniform_std = np.std(uniform_data)

print("Normal Mean:", normal_mean)
print("Uniform Mean:", uniform_mean)

print("Normal Std:", normal_std)
print("Uniform Std:", uniform_std)
```

# 4. Explanation

```python

print("\nExplanation:")

print("Normal distribution is bell-shaped with most values around the mean.")
print("Uniform distribution has equal probability across the range.")

print("\nWhen to use:")

print("Use Normal distribution when data is centered around a mean (e.g., heights, test scores).")
print("Use Uniform distribution when all outcomes are equally likely (e.g., random sampling).")
```
